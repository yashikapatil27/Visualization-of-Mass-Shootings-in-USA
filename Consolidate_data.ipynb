```python
import pandas as pd
import os

directory = 'Mass_Shootings_CSV'
dataframes = []

for year in range(2014, 2023):
    filename = os.path.join(directory, f'{year}.csv')

    if os.path.isfile(filename):
        df = pd.read_csv(filename)
        dataframes.append(df)

appended_data = pd.concat(dataframes, ignore_index=True)
output_csv_file = 'Mass_Shootings_Year-2014_to_2022.csv'


appended_data.to_csv(output_csv_file, index=False)
print(f'Combined data saved to {output_csv_file}')
```
